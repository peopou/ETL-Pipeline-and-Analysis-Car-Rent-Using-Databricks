# 1. Revenue Analysis

```python
%sql
select 
  d.year,
  d.month,
  count(*) as total_transaction,
  sum(f.total_amount) as total_revenue
from gold.fact_rental f
join gold.dim_date d
  on f.rental_date = d.date
group by 1,2
order by 1,2
```

```python
%sql
select
            d.month,
            sum(f.total_amount) as total_revenue
        from gold.fact_rental f
        join gold.dim_date d
            on f.rental_date = d.date
         group by 1
         order by 1
```

# 2. Customer Analysis

```python
%sql
select
  cs.customer_name,
  cs.customer_city,
  cs.customer_age,
  count(*) as rental_count,
  sum(f.total_amount) as total_spent
from gold.fact_rental f
join gold.dim_customer cs
  on f.customer_id = cs.customer_id
group by 1,2,3
order by 5 desc
limit 10
```

```python
%sql
select
  customer_name,
  count(*) as total_rental,
  sum(total_amount) as total_spent
from gold.fact_rental f
join gold.dim_customer cs
  on f.customer_id = cs.customer_id
group by 1
order by 2 desc
limit 10
```

# 3. Vehicle Usage

```python
%sql
select
  v.vehicle_brand,
  v.vehicle_model,
  v.vehicle_type,
  count(*) as total_usage,
  sum(f.rental_duration_days) as total_days
from gold.fact_rental f
join gold.dim_vehicle v
  on f.vehicle_id = v.vehicle_id
group by 1,2,3
order by 4 desc
limit 10
```

# 4. Popular Pickup Location

```python
%sql
select
  l.location_name,
  l.location_city,
  count(*) as total_transaction
from gold.fact_rental f
join gold.dim_location l
  on f.pickup_location_key = l.location_id
group by 1,2
order by 3 desc
```
